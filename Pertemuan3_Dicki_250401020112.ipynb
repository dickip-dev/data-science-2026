```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

df = pd.read_csv("/housing_dirty.csv")
print("Dataset berhasil dimuat!")
print("Shape:", df.shape)
print(df.isnull().sum())
df.info()          # cek tipe data & missing
df.describe()      # statistik dasar
df.isnull().sum()  # jumlah missing per kolom
print("cek duplikat:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print('Setelah hapus duplikat:', df.shape)
```

```python
df['kota']    = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

```python
df['luas_m2']    = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar']      = df['kamar'].fillna(df['kamar'].mode()[0])
```

```python
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)
```

```python
# STEP 5 — Validasi & Ekspor
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```

**Apa yang dipelajari:**
Pertemuan ini mempelajari proses data cleaning meliputi penanganan missing values,
duplikat, standardisasi teks, dan penanganan outlier menggunakan metode IQR clipping
pada dataset properti rumah.

**Temuan utama:**
Dataset awal memiliki 130 baris dengan missing values pada kolom luas_m2 (18),
harga_juta (17), dan kamar (10). Setelah proses cleaning, semua missing values
berhasil ditangani dan dataset bersih tersimpan sebagai housing_clean.csv.
