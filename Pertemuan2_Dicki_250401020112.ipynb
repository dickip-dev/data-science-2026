```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Tampilkan isi
print(df.head())
```

```python
# Pertanyaan 1: Berapa jumlah baris dan kolom?
print(f"Baris: {df.shape[0]}, Kolom: {df.shape[1]}")
print(f"Nama kolom: {df.columns.tolist()}")
```

```python
# Pertanyaan 2 : Kolom mana yang memiliki nilai kosong
print("Jumlah nilai kosong adalah:")
print(df.isnull().sum())
```

```python
# Pertanyaan 3: Berapa persen penumpang selamat?
persen_selamat = (df['Survived'].sum() / len(df) * 100)
print(f"Persentase penumpang selamat adalah: {persen_selamat:.2f}%")
```

```python
# Pertanyaan 4: Penumpang wanita kelas 1
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(f"Jumlah penumpang wanita kelas 1: {len(wanita_kelas1)}")
print(f"Jumlah yang selamat: {wanita_kelas1['Survived'].sum()}")
print(f"Rata-rata usia: {wanita_kelas1['Age'].mean():.2f} tahun")
```

```python
# Pertanyaan 5: Tingkat keselamatan per kelas penumpang
hasil = df.groupby('Pclass').agg(
    total_penumpang=('PassengerId', 'count'),
    total_selamat=('Survived', 'sum'),
    persen_selamat=('Survived', 'mean')
).round(3)

hasil['persen_selamat'] = (hasil['persen_selamat'] * 100).round(2)
print(hasil)

print(f"\n Jadi Kelas dengan tingkat keselamatan tertinggi adalah : Kelas 1 63%")
```

```python
# ④ Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```

**Apa yang dipelajari:**
Pertemuan ini mempelajari eksplorasi data (EDA) menggunakan Pandas pada dataset Titanic yang berisi 891 baris dan 12 kolom, termasuk filtering, groupby, dan agregasi data.

**Temuan utama:**
Sebanyak 38.38% penumpang selamat. Penumpang wanita kelas 1 memiliki tingkat keselamatan tertinggi (91 dari 94 orang selamat). Kelas 1 memiliki tingkat keselamatan 63%,
jauh lebih tinggi dibanding kelas 3 yang hanya 24.2%.
