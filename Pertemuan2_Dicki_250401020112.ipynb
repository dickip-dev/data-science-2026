```python
import pandas as pd
import numpy as np

# Muat dataset Titanic dari URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Tampilkan informasi awal
print("Dataset berhasil dimuat!")
print(df.head())
```

```python
# Q1: Berapa jumlah baris dan kolom?
print(f"Baris: {df.shape[0]}, Kolom: {df.shape[1]}")
print(f"Nama kolom: {df.columns.tolist()}")
```

```python
# Q2: Missing values per kolom
print("Jumlah missing values:")
print(df.isnull().sum())
print("\nProporsi missing (%):")
print((df.isnull().sum() / len(df) * 100).round(2))
```
